```python
import pandas as pd 
import numpy as np
df = pd.read_csv('weather_data.csv')
df
```

```python
new_df = df.replace(-99999,0)
new_df
```

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
new_df = df.replace(to_replace=[-99999,7],value=5)
new_df
```

```python
new_df = df.replace({
    'temperature':-99999,
    'event':'0'
},np.nan)
new_df
```

```python
new_df = df.replace({
    -99999:np.nan,
    '0':'Sunny'
})
new_df
```
